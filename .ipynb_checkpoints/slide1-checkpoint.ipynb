- 数据为拍拍贷真实业务数据，包含328553个信用标信息。
- 包含8个特性：借款金额，借款期限，初始评级，逾期率，借款利率，是否首标，借款类型，借款日期

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
base_color = sb.color_palette()[0]
%matplotlib inline
```

```python
import os
from matplotlib.font_manager import _rebuild
# font_path例子：/opt/conda/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/SimHei.ttf
font_path = os.path.join(plt.__dict__['rcParams']['datapath'],'fonts/ttf/SimHei.ttf')
# 判断字体文件是否存在，不存在则下载
if not os.path.exists(font_path):
    !wget http://d.xiazaiziti.com/en_fonts/fonts/s/SimHei.ttf -O $font_path
# 用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus']=False
# 加载配置
_rebuild()
```

```python
df_lc = pd.read_csv('LC.csv')
df_lc_clean = df_lc.copy()
```

```python
# 保留感兴趣的列
df_lc_clean = df_lc_clean.loc[:,['借款金额','借款期限','借款利率','初始评级','借款成功日期','借款类型','是否首标','历史正常还款期数','历史逾期还款期数']]
# 创建新列: 逾期率 = 正常还款次数 / (正常环框次数 + 逾期还款次数)
df_lc_clean['逾期率'] = df_lc_clean['历史逾期还款期数'] / (df_lc['历史正常还款期数'] + df_lc['历史逾期还款期数'])
# 逾期率所有的nan都是没有开始还款的，将其填充为0
df_lc_clean['逾期率'].fillna(0, inplace=True)
df_lc_clean = df_lc_clean.drop(['历史正常还款期数','历史逾期还款期数'], axis=1)
df_lc_clean['借款成功日期'] = pd.to_datetime(df_lc_clean['借款成功日期'])
```

```python
# 将初始评级变成category类型，并指定顺序
ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = ['A','B','C','D','E','F'])
df_lc_clean['初始评级'] = df_lc_clean['初始评级'].astype(ordered_var)
# 将借款期限按(0-6],(6-12],(12-24]分成category类型
df_lc_clean['借款期限'] = pd.cut(df_lc_clean['借款期限'], [0,6,12,18,24])
# 将借款成功日期分成4份，创建分类列
df_lc_clean['cut_借款成功日期'] = pd.cut(df_lc_clean['借款成功日期'],
                              [pd.Timestamp('2015-01-1 00:00:00'),pd.Timestamp('2015-05-31 23:59:59'),
                               pd.Timestamp('2015-12-31 23:59:59'),pd.Timestamp('2016-5-31 23:59:59'),
                               pd.Timestamp('2017-01-30 00:00:00')],include_lowest=True,
                                  labels=['15年上','15年下','16年上','16年下'])
```

```python
# log变换(10为底)
def trans(x, inverse = False):
    if not inverse: return np.log10(x)
    else: return 10 ** x
# 创建log变换后的数据
df_lc_clean['log_借款金额'] = df_lc_clean['借款金额'].apply(trans)
# 绘制log处理后数据的直方图
bin_edges = np.arange(0.1, df_lc_clean['log_借款金额'].max()+0.3, 0.3)
plt.hist(df_lc_clean['借款金额'].apply(trans), bins = bin_edges)
tick_locs = pd.Series([10, 100, 1000, 10000, 100000, 1000000]).apply(trans)
plt.xticks(tick_locs, [10,100,'1k','10k','100k','1000k']);
```

## Multivariate Exploration

想知道我有借款金额，借款期限，会得到什么样借款利率

### 还有什么影响了借款利率

```python
bin_log_edges = np.arange(trans(100), trans(500000)+0.5, 0.5)
bin_edges = 10 ** bin_log_edges
bin_centers = 10 ** (bin_log_edges + 0.5/2)[:-1]

def my_errorbar(data,label,color):
    data = data[data['借款期限']==label]
    # pd.cut分箱
    data_bins = pd.cut(data['借款金额'], bin_edges, right = False, include_lowest = True)
    # # 计算每个分组的统计量(均值)
    y_means = data['借款利率'].groupby(data_bins).mean()
    y_std = data['借款利率'].groupby(data_bins).std()
    # 绘制汇总数据图
    plt.errorbar(x = bin_centers, y = y_means, color=color, label=label, alpha=0.5);

plt.figure(figsize=[10,6])
label = df_lc_clean['借款期限'].value_counts().index
color = ['r','b','g','y']
for j,k in zip(label, color):
    my_errorbar(data=df_lc_clean, label=j, color=k)

plt.xscale('log')
plt.xticks(bin_edges,np.round(bin_edges/100,1))
plt.title('不同借款期限的金额和利率', fontsize=20);
plt.xlabel('借款金额 x1000', fontsize=16);
plt.ylabel('借款利率均值', fontsize=16)
plt.legend(fontsize=15, loc='upper right');
plt.tick_params(labelsize=13)
```

> 借款期限在1年内借款利率随借款金额有下降趋势

### 探索逾期率，初始评级，借款利率的关系

```python
# 将逾期率分箱
bins = [0, 0.25, 0.5, 0.75, 1]
df_lc_clean['cut_逾期率'] = pd.cut(df_lc_clean['逾期率'], bins, include_lowest=True)
# 绘制每个分箱的评级和借款利率
g = sb.FacetGrid(data = df_lc_clean, col = 'cut_逾期率')
g.map(sb.pointplot, '初始评级', '借款利率', linestyles = "", order=['A','B','C','D','E','F']);
```

> 初始评级和借款利率负相关，然而逾期率对借款利率没有直接影响。
